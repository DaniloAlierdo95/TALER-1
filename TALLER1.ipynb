```python
##EJERCICIO 10


matriz  =  [[62, 42, 70, 66, 10],
            [35, 10,  4, 92, 94],
            [36, 37,  5,100, 19],
            [79, 44, 72, 53, 91]]


m = matriz        # m--> matriz de entrada


## --Preparo las dimensiones de la matriz de salida mo
mo = [None]*len(m[0])          #nro de columnas de m = nro de columnas mo
for i in range(0,len(m[0])):
    mo[i] = [None]*len(m)      #nro de filas de m = nro de columnas de mo

for i in range(len(m)):
    for j in range(-len(m[0]),0): # empiezo al revés para asegurar 
                                    #el giro antihorario
        mo[-j-1][i] = m[i][j]    # indice de la fila no ascendente e indice 
                                #de la columna m descendente
print()
print("Matriz Original") # matriz original
for i in range(len(m)):
    print(m[i])

print()     # matriz rotada 
print("Matriz Rotada 90 grados en sentido antihorario")
for i in range(len(mo)):
    print(mo[i])
```

```python
##EJERCICIO 11


def cuadrilátero(ancho, alto, caracter):

  for i in range(0, alto):
 
   for j in range(0, ancho): #%%ciclo anidado
  
    print(caracter, end="")
  
  print() ##da un salto de linea

 

##definimos las
 
ancho = int(input('ingrese la anchura de la figura: '))

alto = int(input('ingrese la altura de la figura: '))

caracter = input('ingrese el caracter que va aformar la figura: ')


##llamamos a la función cuadrilátero y le enviamos los parámetros

cuadrilátero(ancho, alto, caracter)
```

```python
##EJERCICIO 12

def triangulo(ancho_trian):

  for i in range(1, ancho_trian + 1):##este ciclo anidado va colocando los caracteres que van aumentando
 
   for j in range(i):
 
     print("* ", end = "")
 
   print()

 

 for i in range(1, ancho_trian):##este otro cicloanidado va disminuyendo eltriangulo de lamitad hacia abajo
 
   for j in range(ancho_trian - i):
 
     print("* ", end="")
 
   print()



ancho_trian =int(input('ingrese laanchura deltriángulo: '))

triangulo(ancho_trian)


```

```python
##EJERCICIO 13


##función que define si un año es bisiesto
def
 año_bisiesto(time):

  return time % 400 == 0 or (time % 100 !=0 and time % 4==0)



##ingresamos los años a estudiar
  
print('contador de años bisiestos ')

año_inicial = int(input('escriba un año: '))

print('escriba un año posterior a: ',str(año_inicial)+" : ",end="")

año_final = int(input())



##verificamos si el segundo año es mayor que elprimero


while año_final < año_inicial:
 
 print(año_final, 'no es mayor que ',str(año_inicial) + ", intente un año mayor: ", end = "" )
 
 año_final = int(input()) 



##cuenta los años bisiestos probados en lafunsion


contador = 0 

for i in range(año_inicial, año_final + 1):

  if año_bisiesto(i): 
 
    contador += 1




print("De ",año_inicial," a ",año_final," hay ",año_final - año_inicial + 1," años, ",contador," de ellos son bisiestos")
```
